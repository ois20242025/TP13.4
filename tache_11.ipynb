```r
# Générer des données aléatoires pour la visualisation
set.seed(123)
data <- data.frame(
  x = rnorm(100),
  y = rnorm(100),
  z = rnorm(100)
)

# Créer une visualisation avec ggplot2
ggplot(data, aes(x = x, y = y, color = z)) +
  geom_point(size = 5) +
  scale_color_gradient(low = "blue", high = "red") +
  theme_minimal()
```
